### üîç Finding Outliers and Their Relationship with Target Variable (y) Using IQR Method

```python
for col in num:
    Q1 = bank[col].quantile(0.25)
    Q3 = bank[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    outliers = bank[(bank[col] < lower ) | (bank[col] > upper)]
    #print(col, "outliers :", outliers.shape[0])
    print(col)
    bank['outlier_flag'] = ((bank[col] < lower) | (bank[col] > upper)).astype(int)
    print(bank.groupby('outlier_flag')['y'].value_counts(normalize=True))
```

```python
bank = bank.drop(columns='outlier_flag')
```

```python
Q1 = bank['campaign'].quantile(0.25)
Q3 = bank['campaign'].quantile(0.75)

IQR = Q3-Q1

upper_cap = Q3 + 1.5 * IQR

bank['campaign'] = bank['campaign'].clip(upper=upper_cap) 
```

### üìå Handling Outliers in the Campaign Feature
* ‚ÄúThe campaign feature had many outliers and showed a weak or negative relationship with the target variable, so capping was applied to reduce the impact of extreme values instead of removing data.‚Äù

```python
(bank == "unknown").sum().sort_values(ascending=False)
```

### üìå Replace Unknown Values
* ‚ÄúWe replace unknown values because they indicate missing data rather than real categories, and keeping them can negatively impact both model performance and interpretability.‚Äù

```python
bank.default.replace("unknown", bank['default'].mode()[0], inplace=True)
bank.education.replace("unknown", bank['education'].mode()[0], inplace=True)
bank.housing.replace("unknown", bank['housing'].mode()[0], inplace=True)
bank.loan.replace("unknown", bank['loan'].mode()[0], inplace=True)
bank.job.replace("unknown", bank['job'].mode()[0], inplace=True)
bank.marital.replace("unknown", bank['marital'].mode()[0], inplace=True)
```

### üîÑ Encoding Categorical Features Using LabelEncoder
* ‚ÄúLabelEncoder used to convert categorical features into numerical form, mapping categories to numeric values, with binary features encoded as 0 and 1.‚Äù

```python
le = LabelEncoder()
```

```python
bank[bank.select_dtypes(include='object').columns] = bank[bank.select_dtypes(include='object').columns].apply(le.fit_transform)
```

### üîÄ Train‚ÄìTest Split
* ‚ÄúThe dataset was split into training and testing sets using a stratified 80‚Äì20 split to preserve class distribution and ensure reliable model evaluation.‚Äù

```python

bank_train, bank_test = train_test_split(bank, random_state=42, test_size=.2)
```

```python
bank_train.y.value_counts()
```

### üìå Random Over-Sampling (Manual Duplication)
* ‚ÄúClass imbalance was handled using random over-sampling, where minority class samples were duplicated to balance the dataset.‚Äù

```python
df1 = bank_train[bank_train.y==1]

bank_train = pd.concat([bank_train, df1, df1, df1, df1,  df1, df1, df1])
```

```python
#pip install imbalanced-learn
```

```python
bank_train.y.value_counts()
```

```python
bank_train_x = bank_train.iloc[:,:-1]
bank_train_y = bank_train.iloc[:,-1]
```

```python
bank_test_x = bank_test.iloc[:,:-1]
bank_test_y = bank_test.iloc[:,-1]
```

```python
bank_train_x.shape
```

```python
bank_train_y
```

```python
bank_train_x.isnull().sum()
```

```python
# Initialize LogisticRegression Model
model = LogisticRegression()
```

```python
# Train the LogisticRegression Model
model.fit(bank_train_x, bank_train_y)
```

```python
# Predict on Test data
pred = model.predict(bank_test_x)
```

```python
# Check Accuracy of the Model
accuracy_score(bank_test_y, pred)
```

```python
# Understand the model using conconfusion matrix
confusion_matrix(bank_test_y, pred)
```

```python
# Recall Score
recall_score(bank_test_y, pred)
```

```python
# F1 Score
f1_score(bank_test_y, pred)
```

```python
# classification report of the Model
print(classification_report(bank_test_y, pred))
```

### Logistic Regression Model Observation
* Achieved an accuracy of 84.18%, which improved to 90.81% after class balancing.

* Demonstrated high recall (84.96%) for the positive class (y = 1), meaning it correctly identified most subscribers.

* Precision for the positive class remained low (42%), indicating a higher number of false positives.

* F1-score for subscribers was 0.56, showing an imbalance between precision and recall.

* Performed strongly on the majority class (y = 0) with an F1-score of 0.90.

* Suitable as a baseline and interpretable model.

### Conclusion
* Logistic Regression is effective at identifying potential subscribers but tends to over-predict the positive class, leading to lower precision.
* While it offers strong recall and interpretability, its moderate F1-score makes it less optimal for final deployment in a highly imbalanced dataset.

```python
# classification report
print(classification_report(bank_test_y, predict_rf))
```

# Random Forest Classifier Model Observation
* Achieved a high accuracy of 90.81%, outperforming Logistic Regression.

* Delivered a balanced recall of 63.85% and precision of 60% for the positive class.

* Recorded an improved F1-score of 0.62 for subscribers, indicating better balance.

* Significantly reduced false positives (406 vs 1157 in Logistic Regression).

* Handled non-linear relationships and outliers more effectively.

* Demonstrated stable performance across cross-validation.

# Conclusion
* Random Forest provides a more balanced and reliable model with improved precision‚Äìrecall trade-off and higher F1-score.
* Its robustness to outliers and complex feature interactions makes it the preferred model for production use in bank marketing prediction.
